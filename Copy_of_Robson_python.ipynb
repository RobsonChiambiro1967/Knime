```python
def normalise_rectangle(rect):
  """normalise a rectangle so that it starts at the origin and is a 1 unit input should be (x0,y0,x1,y1).
  input should be of the format (x0, y0, x1,y1)
  (x0,y0 should define the lower corner, (x1,y1) should be the upper right corner
    """
  assert len(rect) == 4,'Rectangles must contain four sides'
  x0, y0, x1, y1 = rect
  assert x0 < x1, 'Invalid x coordinates'
  assert y0 <y1, 'Invalid y coordinates'
  dx = x1 - x0
  dy = y1 - y0
  if dx > dy:
    scaled = dy /dx
    upper_x = 1.0
    upper_y = scaled
  else:
    scaled = dx /dy
    upper_x = scaled
    upper_y = 1.0

  assert 0 < upper_x <= 1.0,'Calculated upper X coordinate invalid'
  assert 0 < upper_y <= 1.0, 'Calculated upper Y coordinate invalid'

  return (0, 0,upper_x, upper_y)

print(normalise_rectangle((1.0, 2.0, 4.0, 7.0)))

```
